```python
import pandas as pd
data=pd.read_pickle('data_cache/stock')
data.head()
```

```python
#get data in level outer is easy
data['Close']

#how to get data in inner part like AAPL on Close, High, and Low?
```

```python
#get data inner, one

data['Close']['AAPL']
```

```python
#get data all

data.xs(key='AAPL', level='Symbols',axis=1)
```

```python
#change column to row
#using stack

data.stack() #automatic for inner
```

```python
data.stack(level="Attributes")
```

```python
#unstack = change row to column
#automatic fill inner
data.stack(level='Attributes').unstack()
```

```python
#melt: column to row for 2dimension data

fb=data.xs(key='FB', level='Symbols',axis=1)
fb.melt()
```

```python
#reset date not to be index 

fb2=fb.reset_index().melt(id_vars='Date')
fb2
```

```python
#sortir attributes

fb_=fb.reset_index().melt(id_vars='Date',value_vars=['Close','Low'],
                      var_name='att',value_name='value')
fb_
```

```python
fb
```

```python
#groupby for data 2 dimension

d=fb2.groupby('Attributes').mean().sort_values(by='value',ascending=False)
d
```

```python
import matplotlib.pyplot as plt
print(plt.style.available)
plt.style.use('seaborn')
```

```python
d[1:].plot(kind='line')
```

```python
data['Volume'][['AAPL']].plot(kind='box',vert=False) #vert make horizontal
```

```python
norm=data['Close'].std()/data['Close'].mean()
norm.plot(kind='bar')
```

```python
#combine make data to quarter, month name, day name

closingprice=data['Close']
closingprice['Month']=closingprice.index.month_name()
average_month=closingprice.groupby('Month').mean()
average_month
```

```python
average_month.plot()
```

```python
average_month.plot(kind='bar')
```

```python
quarter1=['January', 'February', 'March']
average_month.reindex(quarter1).plot(kind='bar')
```

```python
#to quarter

volum_apple=data['Volume']['AAPL'].reset_index()
volum_apple['Q']=volum_apple['Date'].dt.to_period('Q')
volum_apple
```

```python
volum_apple.boxplot(column='AAPL',by='Q',vert=False)
```
